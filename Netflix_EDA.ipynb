## 🔹 Data Cleaning
Steps performed:
1. Handle missing values
2. Convert `date_added` into datetime
3. Extract useful features (`year_added`, `content_age`)
4. Remove duplicates


```python
# df=df.drop(['season_num','movie_duration'],axis=1)
```

```python
df=df.drop('duration',axis=1)
```

```python
df.isnull().sum()

df=df.fillna({"director":"Unknown",'cast':"Unknown","country":"Unknown",'genres':"Unknown","description":"No Description Available"})

df['budget']=pd.to_numeric(df['budget'])

df['revenue']=pd.to_numeric(df['revenue'])
```

For TV Shows there are no such columns like revenue and budget. but in movies it is there.so there NaNs in tv shows.


```python
profit=df['revenue']-df['budget']
ROI=profit/df['budget']
df['profit']=profit
df['ROI']=ROI
# df.head(15)
```

```python
df['ROI']=df['ROI'].replace([np.inf,-np.inf],np.nan)
```

```python
# df['loss']=df['loss'].fillna(np.nan)
```

**Profit and ROI apply only to Movies**

```python
profit_mean=df['profit'].mean()
loss = df['profit']<0
df['loss']=loss
```
