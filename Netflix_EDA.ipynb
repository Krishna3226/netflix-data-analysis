## 🔹 Data Cleaning
Steps performed:
1. Handle missing values
2. Convert `date_added` into datetime, Extract useful feature `content_age`
3. Remove duplicates


## 1. Handle missing values



```python
# df=df.drop(['season_num','movie_duration'],axis=1)
```

```python
df_sorted=df_sorted.drop('duration',axis=1)
```

```python
df_sorted.isnull().sum()

df_sorted=df_sorted.fillna({"director":"Unknown",'cast':"Unknown","country":"Unknown",'genres':"Unknown","description":"No Description Available"})
```

```python
df_sorted['budget']=pd.to_numeric(df_sorted['budget'])

df_sorted['revenue']=pd.to_numeric(df_sorted['revenue'])
```


## 2. Convert `date_added` into datetime,Extract useful feature `year_added`, `content_age`


```python
df_sorted['date_added']=pd.to_datetime(df_sorted['date_added'],errors='coerce')
df_sorted.info()
```

```python
df_sorted['year_added']=df_sorted['date_added'].dt.year
```

```python
df_sorted['content_age']=2026-df_sorted['year_added']
```

```python
display(df_sorted[['year_added','content_age']])
```

## 3. Remove duplicates

```python
duplicates=df_sorted.duplicated().sum()
duplicates
```

For TV Shows there are no such columns like revenue and budget. but in movies it is there.so there NaNs in tv shows.


```python
profit=df_sorted['revenue']-df_sorted['budget']
ROI=profit/df_sorted['budget']
df_sorted['profit']=profit
df_sorted['ROI']=ROI
# df_sorted.head(15)
```

```python
df_sorted['ROI']=df_sorted['ROI'].replace([np.inf,-np.inf],np.nan)
```

**Profit and ROI apply only to Movies**

```python
profit_mean=df_sorted['profit'].mean()
loss = df_sorted['profit']<0
df_sorted['loss']=loss
```

```python
df_sorted[df_sorted['loss'] == True]
```

**Standard boolean columns in pandas (dtype bool) do not support NaN values. If you try to insert np.nan into a standard boolean column, pandas will usually convert the entire column to object dtype or float**

Converting 'loss' Data Type from 'bool' to 'boolean' dtype. This allows the column to hold True, False, and pd.NA (the pandas-specific version of NaN for non-floats).

```python
df_sorted['loss']=df_sorted['loss'].astype('boolean')
```

```python
df_sorted.loc[df_sorted['type']=='TV Show','loss']=pd.NA
```
