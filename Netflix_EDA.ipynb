## 🔹 Data Cleaning
Steps performed:
1. Handle missing values
2. Convert `date_added` into datetime
3. Extract useful features (`year_added`, `content_age`)
4. Remove duplicates


```python
df.isnull().sum()

df.fillna({"director":"Unknown",'cast':"Unknown","country":"Unknown",'genres':"Unknown","description":"No Description Available"})

split_duration=df['duration'].str.split(' ',expand=True)

season_num=np.where(df['type']=='TV Show',split_duration[0],np.nan)
df['season_num']=pd.to_numeric(season_num,errors='coerce')

movie_duration=np.where(df['type']=='Movie',split_duration[0],np.nan)
df['movie_duration']=pd.to_numeric(movie_duration,errors='coerce')

df['budget']=pd.to_numeric(df['budget'])

df['revenue']=pd.to_numeric(df['revenue'])
```

For TV Shows there are no such columns like revenue and budget. but in movies it is there.so there NaNs in tv shows.


```python
profit=df['revenue']-df['budget']
ROI=profit/df['budget']
df['profit']=profit
df['ROI']=ROI
df.head(15)
```

```python
df['ROI']=df['ROI'].replace([np.inf,-np.inf],np.nan)
```
